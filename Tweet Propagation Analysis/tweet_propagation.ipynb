```python
def days_between(d1, d2):
    d1 = datetime.datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.datetime.strptime(d2, "%Y-%m-%d")
    return abs((d2 - d1).days)
    
def users_activity (users_data , followers = True , friends = True , favourites = True ):
    #####normal accounts
    mask_normal = users_data['verification'] == False
    normal_people = users_data[mask_normal]
    activity = np.zeros(len(list(normal_people['user_ID'])))
    if followers:

        activity += normal_people['follower_count'] / max(list(normal_people['follower_count']))
    if friends : 
        activity += normal_people['friends_count'] / max(list(normal_people['friends_count']))
    if favourites :
        activity += normal_people['favaurites_count'] / max(list(normal_people['favaurites_count']))
    activity += normal_people['posts_per_day'] / max(list(normal_people['posts_per_day']))

    normal_people['activity']  = activity / 4 
    activity = activity / max(activity)
    normal_people['normalized activity']  = activity
    
    #####verified accounts
    mask_verification = users_data['verification'] == True
    verified_people = users_data[mask_verification]
    activity = np.zeros(len(list(verified_people['user_ID'])))
    
    if followers:
        activity += verified_people['follower_count'] / max(list(verified_people['follower_count']))
    activity += verified_people['posts_per_day'] / max(list(verified_people['posts_per_day']))

    verified_people['activity'] = activity / 2
    activity = activity / max(activity)
    verified_people['normalized activity'] = activity

    ####merging normal people and verified people
    frames = [verified_people, normal_people]
    acting_data = pd.concat(frames)
    acting_data = acting_data.sort_values(by=['cascade'])
    acting_data = acting_data.reset_index(drop=True)
    
    return acting_data

```

```python
users = list(data['user_ID'])
num_of_users = len(users)
verification = np.zeros(num_of_users)
follower_count = np.zeros(num_of_users)
friends_count = np.zeros(num_of_users)
favaurites_count = np.zeros(num_of_users)
posts_per_day = np.zeros(num_of_users)
today = str(datetime.date.today())
# print(num_of_users)

for i in range(num_of_users):
    print(i)
    try:
        user_obj = api.get_user(users[i])
        verification[i] = user_obj.verified
        follower_count[i] = user_obj.followers_count
        friends_count[i] = user_obj.friends_count 
        favaurites_count[i] = user_obj.favourites_count

        n_posts = user_obj.statuses_count
        creation_time = str(user_obj.created_at)
        creation_time = creation_time[:creation_time.find(' ')]
        duration = days_between(today, creation_time)
        posts_per_day[i] = n_posts / duration                                  
    except tweepy.error.TweepError:
        pass
    

 
```

```python
data['follower_count'] = follower_count
data['friends_count'] = friends_count
data['favaurites_count'] = favaurites_count
data['posts_per_day'] = posts_per_day
data['verification'] = verification
activation = users_activity(data)
```
